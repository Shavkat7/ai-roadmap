```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r"D:\airepos\forked_ai_roadmap\ai-roadmap\content\ml\supervised\02 multiple linear regression\data\housing.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
x = df['area'].values
y = df['price'].values
```

```python
plt.scatter(x, y)
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
beta_1 = (np.mean(x * y) - np.mean(x) * np.mean(y))/(np.mean(x**2) - np.mean(x)**2)
beta_0 = np.mean(y) - beta_1 * np.mean(x)
```

```python
beta_1, beta_0
```

```python
y_pred = beta_0 + beta_1 * x
```

```python
plt.scatter(x, y)
plt.xlabel("Area")
plt.ylabel("Price")
plt.plot(x, y_pred, color='red')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x.reshape(-1, 1), y)
```

```python
predicted_y = lin_reg.predict(x.reshape(-1, 1))
```

```python
plt.scatter(x, y)
plt.xlabel("Area")
plt.ylabel("Price")
plt.plot(x, predicted_y, color='black')
plt.show()
```
